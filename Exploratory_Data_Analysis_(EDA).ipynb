```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
```

```python
churn_rate = df['Churn'].value_counts(normalize=True) * 100
plt.figure(figsize=(6,4))
sns.barplot(x=churn_rate.index, y=churn_rate.values, palette="coolwarm")
plt.title("Churn Rate")
plt.ylabel("Percentage")
plt.xlabel("Churn")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.countplot(data=df, x='gender', hue='Churn', palette="pastel", ax=axes[0])
axes[0].set_title("Churn by Gender")
sns.countplot(data=df, x='SeniorCitizen', hue='Churn', palette="pastel", ax=axes[1])
axes[1].set_title("Churn by Senior Citizen Status")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['tenure'], kde=True, bins=30, color='blue')
plt.title("Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.ylabel("Count")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.countplot(data=df, x='Contract', hue='Churn', palette="viridis", ax=axes[0])
axes[0].set_title("Churn by Contract Type")
sns.countplot(data=df, x='PaymentMethod', hue='Churn', palette="magma", ax=axes[1])
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')
axes[1].set_title("Churn by Payment Method")
plt.tight_layout()
plt.show()
```
