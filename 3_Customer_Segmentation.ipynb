```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
```

```python
segmentation_features = df[['tenure', 'MonthlyCharges']]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Segment'] = kmeans.fit_predict(segmentation_features)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Segment', hue='Churn', palette="coolwarm")
plt.title("Churn Rate by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='tenure', y='MonthlyCharges', hue='Segment', palette='viridis')
plt.title("Customer Segments based on Tenure and Monthly Charges")
plt.xlabel("Tenure (Months)")
plt.ylabel("Monthly Charges ($)")
plt.show()
```
